## Enabling Observability and Logging

```python
import os
import asyncio
from agent_framework.azure import AzureOpenAIChatClient
from azure.identity import AzureCliCredential
```

```python
from dotenv import load_dotenv

load_dotenv('.env')
OPENAI_API_ENDPOINT = os.getenv("OPENAI_API_ENDPOINT")
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
```

### Using OpenTelemetry to set up logging
Required OpenTelemetry dependencies are installed by default in the Agent Framework. There is a `setup_observability` function that configures OpenTelemetry with sensible defaults. By default, it exports to the console if no specific exporter is configured.

See [this page](https://learn.microsoft.com/en-us/agent-framework/tutorials/agents/enable-observability?pivots=programming-language-python) for additional details on OpenTelemetry parameters that can be set in `setup_observability`.

```python
from agent_framework.observability import setup_observability

# Enable agent framework telemetry with console output (default behavior)
setup_observability(enable_sensitive_data=True)
```

```python
# Now we can create an agent and run it; notice the additional telemetry output in the console:
from agent_framework import ChatAgent

bad_joke_agent = ChatAgent(
    chat_client=AzureOpenAIChatClient(
        deployment_name="gpt-5-mini", endpoint=OPENAI_API_ENDPOINT, api_key=OPENAI_API_KEY
    ),
    name="Bad Joker",
    instructions="You try hard but are very bad at telling jokes."
)

result = await bad_joke_agent.run("Tell me a joke about a pirate.")
print(result.text)
```

### AI Foundry integration
If using AI Foundry, we can use the `setup_azure_ai_observability` method, which retrieves the Application Insights connection string from your Azure AI Foundry project and calls setup_observability automatically.

```python
from agent_framework.azure import AzureAIAgentClient

agent_client = AzureAIAgentClient(
    credential=AzureCliCredential(),
    project_endpoint="https://<your-project>.foundry.azure.com"
)

# Automatically configures observability with Application Insights
await agent_client.setup_azure_ai_observability()
```
