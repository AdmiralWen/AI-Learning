```python
import os
import openai
from dotenv import load_dotenv
from azure.identity import AzureCliCredential, get_bearer_token_provider
```

```python
# Load env variables:
load_dotenv('.env')
OPENAI_API_ENDPOINT=os.getenv("OPENAI_API_ENDPOINT")
AZURE_OPENAI_DEPLOYMENT_NAME=os.getenv("AZURE_OPENAI_DEPLOYMENT_NAME_52")
OPENAI_API_VERSION=os.getenv("OPENAI_API_VERSION")
RESOURCE_TENANT_ID=os.getenv("RESOURCE_TENANT_ID")
```

```python
# Make sure to have the azure cli installed.
# In terminal type `az login` to login, then login on your browser popup, then select an environment in the terminal (usually 1):
credential = AzureCliCredential(tenant_id=RESOURCE_TENANT_ID)
token_provider = get_bearer_token_provider(credential, "https://cognitiveservices.azure.com/.default")
```

```python
# Create OpenAI client:
client = openai.AzureOpenAI(
    azure_endpoint = OPENAI_API_ENDPOINT,
    api_version = OPENAI_API_VERSION,
    azure_ad_token_provider=token_provider
)
```

```python
# Test chat completion:
response = client.chat.completions.create(
    model=AZURE_OPENAI_DEPLOYMENT_NAME,
    messages=[
        {"role": "system", "content": "You're a helpful assistant."},
        {"role": "user", "content": "Explain what MCP servers are in simple terms."}
    ]
)
print(response.choices[0].message.content)
```
